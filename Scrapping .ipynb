```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
li="https://www.jobberman.com/jobs"
re=requests.get(li)
re
```

```python
data=BeautifulSoup(re.content,"lxml")
print(data.prettify())
```

```python
job_name=[]
for i in data.find_all("p", class_="text-lg font-medium break-words text-link-500"):
    job_name.append(i.text.strip())
print(job_name)
len(job_name)
```

```python
describe=[]
for h in data.find_all("p", class_="text-sm font-normal text-gray-700 md:text-gray-500 md:pl-5"):
    describe.append(h.text.strip())
print(describe)
len(describe)
```

```python
post=[]
for k in data.find_all("p", class_="ml-auto text-sm font-normal text-gray-700 text-loading-animate"):
    post.append(k.text.strip())
print(post)
len(post)
```

```python
company=[]
for l in data.find_all("a", class_="text-loading-animate text-loading-animate-link"):
    company.append(l.text.strip())
print(company)
len(company)
```

```python
op=[]
for h in data.find_all("span",class_="mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide"):
    op.append(h.text.split())
print(op)
w=op[1::3]
#print(w)
r=op[0::3]
#print(r)
```

```python
p=[]
for h1 in data.find_all("p",class_="text-sm text-gray-500 text-loading-animate inline-block"):
    p.append(h1.text.split())
print(p)
len(p)
```

```python
l1,l2,l3,l4,l5,l6,l7,l8,l9=[],[],[],[],[],[],[],[],[]
for page in range(1,15):
    link=f"https://www.jobberman.com/jobs?page={page}"
    re=requests.get(link)
    data=BeautifulSoup(re.content,"lxml")
    role=data.find_all("p", class_="text-lg font-medium break-words text-link-500")
    job_function=data.find_all("p",class_="text-sm text-gray-500 text-loading-animate inline-block")
    company=data.find_all("a", class_="text-loading-animate text-loading-animate-link")
    work_type=data.find_all("span",class_="mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide")
    post=data.find_all("p", class_="ml-auto text-sm font-normal text-gray-700 text-loading-animate")
    details=data.find_all("p", class_="text-sm font-normal text-gray-700 md:text-gray-500 md:pl-5")
    
    max_length=max(len(role),len(job_function),len(company),len(post),len(details))
    
    for i in range(max_length):
        if i >= len(role) or not role[i].text.strip():  
            continue 
        l1.append(role[i].text.strip()) 
        l2.append(job_function[i].text.replace("Job Function", "").replace(":", "").strip() if i < len(job_function) else "N/A")
        #l3.append(work_type[i].text.strip() if i < len(work_type) else "N/A")
        l4.append(company[i].text.strip() if i < len(company) else "N/A")
        l5.append(details[i].text.strip() if i < len(details) else "N/A")
        l6.append(post[i].text.strip() if i < len(post) else "N/A")
    for item in work_type:
        l3.append(item.text.strip())

for i in range(0, len(l3), 3):
    l7.append(l3[i] if i < len(l3) else "N/A")       
    l8.append(l3[i + 1] if i + 1 < len(l3) else "N/A") 
    #l9.append(l3[i + 2].replace("NGN", "").replace("\n", "").replace("\t", "").strip() if i + 2 < len(l3) else "N/A")
    if i + 2 < len(l3):
        salary = l3[i + 2].replace("NGN", "").replace("\n", "").replace("\t", "").strip()
        salary = " ".join(salary.split())
        if "Confidential" in salary:
            l9.append("Confidential")
        else:
            l9.append(salary if salary else "N/A")
    else:
        l9.append("N/A")  
print("Scraping completed.")
```

```python
file = pd.DataFrame({"Job_Role" :l1 ,"Company_Name" : l4,"City" :l7,"Work_type":l8,"Salary":l9, 'Job_function':l2,"Description" : l5 , "Post_Time" : l6 })
file
```

```python
file.to_csv("jobberman_jobs.csv", index=False)

print("CSV file created: jobberman_jobs.csv")
```
